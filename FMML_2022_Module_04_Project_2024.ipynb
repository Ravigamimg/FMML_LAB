## Slope & Derivatives

Creating a function for the derivative of $f(x)$ called df(x)

```python
def df(x):
    return 2 * x + 1
```

```python
# Plot function and derivative side by side
plt.figure(figsize=[15, 5])

# 1 Chart: Cost function
plt.subplot(1, 2, 1)

plt.xlim(-3, 3)
plt.ylim(0, 8)

plt.title('Cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)

plt.plot(x_1, f(x_1), color='blue', linewidth=3)

# 2 Chart: Derivative
plt.subplot(1, 2, 2)

plt.title('Slope of the cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.grid()
plt.xlim(-2, 3)
plt.ylim(-3, 6)

plt.plot(x_1, df(x_1), color='skyblue', linewidth=5)

plt.show()
```

## Gradient Descent

```python
##########################
##########################

## TASK-1 : Complete the lines of code wherever marked as [REQUIRED] in this cell.

##########################
##########################

# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multiplier * gradient

    step_size = abs(new_x - previous_x)
    # print(step_size)

    x_list.append(new_x)
    slope_list.append(df(new_x))

    if step_size < precision:
        print('Loop ran this many times:', n)
        break

print('Local minimum occurs at:', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('f(x) value or cost at this point is:', f(new_x))
```

```python
# Superimpose the gradient descent calculations on plot

plt.figure(figsize=[20, 5])

# 1 Chart: Cost function
plt.subplot(1, 3, 1)

plt.xlim(-3, 3)
plt.ylim(0, 8)

plt.title('Cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)

plt.plot(x_1, f(x_1), color='blue', linewidth=3, alpha=0.8)

values = np.array(x_list)
plt.scatter(x_list, f(values), color='red', s=100, alpha=0.6)

# 2 Chart: Derivative
plt.subplot(1, 3, 2)

plt.title('Slope of the cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.grid()
plt.xlim(-2, 3)
plt.ylim(-3, 6)

plt.plot(x_1, df(x_1), color='skyblue', linewidth=5, alpha=0.6)
plt.scatter(x_list, slope_list, color='red', s=100, alpha=0.5)

# 3 Chart: Derivative (Close Up)
plt.subplot(1, 3, 3)

plt.title('Gradient Descent (close up)', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.grid()
plt.xlim(-0.55, -0.2)
plt.ylim(-0.3, 0.8)

plt.plot(x_1, df(x_1), color='skyblue', linewidth=6, alpha=0.8)
plt.scatter(x_list, slope_list, color='red', s=300, alpha=0.6)


plt.show()
```

## Gradient Descent

```python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001,
                    max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list
```

```python
## TASK-7 : Plot reduction in cost for each iteration
# Function and its derivative
def g(x):
    return x**2 + 3*x + 2

def dg(x):
    return 2*x + 3

# Gradient Descent Function
def gradient_descent_with_cost(derivative_func, initial_guess, multiplier, precision, max_iter):
    x = initial_guess
    x_list = [x]
    cost_list = [g(x)]

    for _ in range(max_iter):
        slope = derivative_func(x)
        x = x - multiplier * slope
        x_list.append(x)

        cost = g(x)
        cost_list.append(cost)

        if abs(x_list[-1] - x_list[-2]) < precision:
            break

    return x_list, cost_list

# Run gradient descent for different learning rates with cost tracking
n = 100

# Low Gamma
low_gamma, cost_low_gamma = gradient_descent_with_cost(derivative_func=dg, initial_guess=3, multiplier=0.0005, precision=0.0001, max_iter=n)

# Mid Gamma
mid_gamma, cost_mid_gamma = gradient_descent_with_cost(derivative_func=dg, initial_guess=3, multiplier=0.001, precision=0.0001, max_iter=n)

# High Gamma
high_gamma, cost_high_gamma = gradient_descent_with_cost(derivative_func=dg, initial_guess=3, multiplier=0.002, precision=0.0001, max_iter=n)

# Insane Gamma
insane_gamma, cost_insane_gamma = gradient_descent_with_cost(derivative_func=dg, initial_guess=1.9, multiplier=0.25, precision=0.0001, max_iter=n)

# Plot the results
plt.figure(figsize=(16, 8))

# Plot reduction in cost for each iteration
plt.subplot(2, 2, 1)
plt.plot(cost_low_gamma, label='Low Gamma')
plt.title('Reduction in Cost - Low Gamma')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(cost_mid_gamma, label='Mid Gamma')
plt.title('Reduction in Cost - Mid Gamma')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(cost_high_gamma, label='High Gamma')
plt.title('Reduction in Cost - High Gamma')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(cost_insane_gamma, label='Insane Gamma')
plt.title('Reduction in Cost - Insane Gamma')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.legend()

plt.tight_layout()
plt.show()
```

Minimise $$f(x, y) = \frac{1}{r + 1}$$ where $r$ is $3^{-x^2 - y^2}$

```python
## TASK-8 : Code the same things as they were done for the first two examples
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Function and its derivative
def f1(x, y):
    return 13 - x**2 - y**2 + 1

def df1_dx(x, y):
    return -2 * x

def df1_dy(x, y):
    return -2 * y

def f2(x, y):
    r = 3 - x**2 - y**2
    return 1 / (r + 1)

def df2_dx(x, y):
    r = 3 - x**2 - y**2
    return 2 * x / (r + 1)**2

def df2_dy(x, y):
    r = 3 - x**2 - y**2
    return 2 * y / (r + 1)**2

# Gradient Descent Function
def gradient_descent_with_cost_2d(derivative_x, derivative_y, initial_guess, multiplier, precision, max_iter, func):
    x, y = initial_guess
    x_list, y_list, cost_list = [x], [y], [func(x, y)]

    for _ in range(max_iter):
        slope_x = derivative_x(x, y)
        slope_y = derivative_y(x, y)
        x = x - multiplier * slope_x
        y = y - multiplier * slope_y
        x_list.append(x)
        y_list.append(y)

        cost = func(x, y)
        cost_list.append(cost)

        if abs(cost_list[-1] - cost_list[-2]) < precision:
            break

    return x_list, y_list, cost_list

# Run gradient descent for the first function
initial_guess = (2, 2)
n = 100

# Gradient Descent for the first function
x1, y1, cost1 = gradient_descent_with_cost_2d(df1_dx, df1_dy, initial_guess, 0.1, 0.0001, n, f1)

# Plot the results for the first function
fig = plt.figure(figsize=(12, 6))

# Plot the function surface
ax1 = fig.add_subplot(121, projection='3d')
x_range = np.linspace(-5, 5, 100)
y_range = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_range, y_range)
Z = f1(X, Y)
ax1.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
ax1.scatter(x1, y1, cost1, color='red', label='Path of Gradient Descent')
ax1.set_title('Function Surface - $f(x, y) = 13 - x^2 - y^2 + 1$')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('f(x, y)')
ax1.legend()

# Plot the reduction in cost for each iteration
ax2 = fig.add_subplot(122)
cost1_np = np.array(cost1)
ax2.plot(cost1_np[:-1] - cost1_np[1:], marker='o', linestyle='-', color='blue', label='Reduction in Cost')
ax2.set_title('Reduction in Cost - $f(x, y) = 13 - x^2 - y^2 + 1$')
ax2.set_xlabel('Iterations')
ax2.set_ylabel('Reduction in Cost')
ax2.legend()

plt.tight_layout()
plt.show()
```

## Mean Squared Error: a cost function for regression problems

### $$RSS = \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$
### $$MSE = \frac{1}{n} \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$
### $$MSE = \frac{1}{n} \sum_{i=1}^{n} \big( y - \hat{y} \big)^2 $$

```python
# Challenge: Write a python function mse(y, y_hat) returns the MSE? Call the mse(y, y_hat)
# function and print out the MSE for the y_hat calculated above.

def mse(y, y_hat):
    #mse_calc = 1/7 * sum((y - y_hat)**2)
    #mse_calc = (1/y.size) * sum((y - y_hat)**2)
    mse_calc = np.average((y - y_hat)**2, axis=0)
    return mse_calc
```

## Partial Derivatives of MSE w.r.t. $\theta_0$ and $\theta_1$

## $$\frac{\partial MSE}{\partial \theta_0} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

## $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

## MSE & Gradient Descent

```python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
```

```python
multiplier = 0.01
thetas = np.array([2.9, 2.9])

# Collect data points for scatter plot
plot_vals = thetas.reshape(1, 2)
mse_vals = mse(y_5, thetas[0] + thetas[1]*x_5)

for i in range(1000):
    thetas = thetas - multiplier * grad(x_5, y_5, thetas)

    # Append the new values to our numpy arrays
    plot_vals = np.concatenate((plot_vals, thetas.reshape(1, 2)), axis=0)
    mse_vals = np.append(arr=mse_vals, values=mse(y_5, thetas[0] + thetas[1]*x_5))

# Results
print('Min occurs at Theta 0:', thetas[0])
print('Min occurs at Theta 1:', thetas[1])
print('MSE is:', mse(y_5, thetas[0] + thetas[1]*x_5))
```
